```python
import pandas as pd
import numpy as np

stud_dict = {"name": ["North", "Nox", "Gryphon", "Atlas", "Gabe"], "age": [28, 25, 26, 21, 21], "marks": [np.nan, 99.48, 91.54, np.nan, 87.34]}

# Create DataFrame from dict
stud_df = pd.DataFrame(stud_dict)
print(stud_df)
```

```python
# drop column with NaN
stud_df = stud_df.dropna(axis='columns')

print(stud_df)
```

```python
# drop column with NaN
stud_df = stud_df.dropna(axis='rows')

print(stud_df)
```

```python
# keep column with 3 or more non-NA values
import pandas as pd
import numpy as np

stud1_dict = {"name": ["North", "Nox", "Gryphon", "Atlas", "Gabe"], "age": [28, np.nan, 26, np.nan, np.nan], "marks": [np.nan, 99.48, 91.54, np.nan, 87.34]}

stud1_df = pd.DataFrame(stud1_dict)
print(stud1_df)
stud1_df = stud1_df.dropna(axis='columns', thresh=3)

print(stud1_df)
```

```python
# drop duplicate rows
import pandas as pd
import numpy as np

stud2_dict = {"name": ["North", "Nox", "Gabe", "Atlas", "North"], "age": [28, np.nan, 26, np.nan, 28], "marks": [np.nan, 99.48, 91.54, 87.34, np.nan]}

stud2_df = pd.DataFrame(stud2_dict)
print(stud2_df)

stud2_df = stud2_df.drop_duplicates()

print(stud2_df)
```

```python
import numpy as np
from sklearn.impute import KNNImputer
nan = np.nan
X = [[1, 2, nan], [3, 4, 3], [nan, 6, 5], [8, 8, 7]]
imputer = KNNImputer(n_neighbors=2, weights="uniform")
imputer.fit_transform(X)
```

```python
from sklearn.impute import SimpleImputer
#Keeping the number of features constant
imputer = SimpleImputer()
X = np.array([[np.nan, 1], [np.nan, 2], [np.nan, 3]])
imputer.fit_transform(X)
```

The SimpleImputer class also supports categorical data represented as string values or pandas categoricals when using the 'most_frequent' or 'constant' strategy:

```python
import pandas as pd
df = pd.DataFrame([["1", "x"],
                   [np.nan, "y"],
                   ["1", np.nan],
                   ["2", "y"]], dtype="category")
imp = SimpleImputer(strategy="most_frequent")
print(imp.fit_transform(df))
```
